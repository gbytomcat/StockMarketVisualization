```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%pylab inline
```

```python
from datetime import datetime
```

```python
"""
For this assignment, 2 financial data sets were used. 
1. The S&P500 Index
2. The VIX S&P500 Index
"""
sp500 = pd.read_csv("http://www.quandl.com/api/v1/datasets/YAHOO/INDEX_GSPC.csv?trim_start=1990-01-02&trim_end=2014-09-19", 
                    parse_dates = ['Date'], dayfirst = True, index_col = 'Date')
vix = pd.read_csv("http://www.quandl.com/api/v1/datasets/YAHOO/INDEX_VIX.csv?trim_start=1990-01-02&trim_end=2014-09-19",
                  parse_dates = ['Date'], dayfirst = True, index_col = 'Date')
```

```python
sp500.head()
```

```python
fig = plt.figure(figsize = (20, 10))

ax1 = fig.add_subplot(2, 1, 1)
ax1.grid(True)
ax1.set_title("S&P 500 variations")
ax1.set_xlabel("Time")
ax1.set_ylabel("Price in USD")
ax1.plot(sp500.index, sp500.Close, 'b')              

ax2 = fig.add_subplot(2, 1, 2)
ax2.set_title("VIX variations")
ax2.set_xlabel("Time")
ax2.set_ylabel("Price in USD")
ax2.grid(True)
ax2.plot(vix.index, vix.Close, 'g')
```
