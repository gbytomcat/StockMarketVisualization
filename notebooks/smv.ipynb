**Getting Started** : 
Load the `default` libraries and set default values for `matplotlib`

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

Downloaded historical Quotes (daily prices: Open, Close, High, Low, Volume) for these 20 odd random stock symbols as a starting point. Load these values to a data frame (examining the Closing prices). 
Github link for [data](https://github.com/pmadan/StockMarketVisualization/tree/master/data)


```python
# \todo need to pick more symbols to find "co-integration" patterns.

symbols = ['AAPL', 'AMZN', 'BAC', 'C', 'F', 'GE', 'GOOG', 'INTC', 'JPM', 'KO', 'MA', 'MRK', 
           'MS', 'MSFT', 'PG', 'V', 'WFC', 'XOM']

# place holder data frames
close = pd.DataFrame()
volume = pd.DataFrame()

# Loop through all the symbols in the list and create a new data frame for daily closing prices 
# for the time period "Jan 01, 2011 to Today"
# \todo  need to adjust different time periods from 2002 onwards to Today.

for s in symbols:
    file_path = "../data/" + s + ".csv"
    d = pd.read_csv(file_path,  parse_dates = ['Date'], dayfirst = True, index_col = 'Date')
    close[s] = d['Close'][d.index > dt.datetime(2010, 12, 31)]
    volume[s] = d['Volume'][d.index > dt.datetime(2010, 12, 31)]

# data to upload to github    
#close.to_csv("../data/Close.csv")
#volume.to_csv("../data/Volume.csv")
```

```python
data = close             # make a local copy of the data frame
data.info()              # what does this data frame have?
```

##Goals 
The goal is to find which stocks vary together. Do they vary together? What measure defines this behaviour? 

To start with, lets pick stocks from the same sector and plot them, assuming that their daily closing price variations might
exhibit similar trends. 

```python
# Basic plots of select columns
fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.grid(True)
ax1.set_title("Stock price variations", fontsize = 25)
ax1.set_xlabel("Time")
ax1.set_ylabel("Price in USD")

# basic plots for 4 bank stocks - 
# JPM - JP Morgan, BAC - Bank of America, C - Citibank, WFC - Wells Fargo
ax1.plot(data.index, data.JPM, label = "JPM", c = 'b')
ax1.plot(data.index, data.BAC, label = "BAC", c = 'g')
ax1.plot(data.index, data.C, label = "C", c = 'r')
ax1.plot(data.index, data.WFC, label = "WFC", c = 'm')
ax1.legend()
```

```python
# Plot the Scatter matrix for the same stocks - JPM, C, BAC, WFC
from pandas.tools.plotting import scatter_matrix
scatter_mat_data = data[["JPM", "C", "BAC", "WFC"]]       # isolate the 4 columns to a new data frame
print scatter_matrix(scatter_mat_data, alpha = 0.5, diagonal = 'kde')
```
