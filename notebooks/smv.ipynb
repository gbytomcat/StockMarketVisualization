```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
# downloaded historical Quotes (daily prices: Open, Close, High, Low, Volume) for these 20 odd random 
# stock symbols as a starting point
# \todo need to pick more symbols to find "co-integration" patterns.

symbols = ['AAPL', 'AMZN', 'BAC', 'C', 'F', 'GE', 'GOOG', 'INTC', 'JPM', 'KO', 'MA', 'MRK', 
           'MS', 'MSFT', 'PG', 'V', 'WFC', 'XOM']

# place holder data frames
close = pd.DataFrame()
volume = pd.DataFrame()

# Loop through all the symbols in the list and create a new data frame for daily closing prices 
# for the time period "Jan 01, 2011 to Today"
# \todo  need to adjust different time periods from 2002 onwards to Today.

for s in symbols:
    file_path = "../data/" + s + ".csv"
    d = pd.read_csv(file_path,  parse_dates = ['Date'], dayfirst = True, index_col = 'Date')
    close[s] = d['Close'][d.index > dt.datetime(2010, 12, 31)]
    volume[s] = d['Volume'][d.index > dt.datetime(2010, 12, 31)]

# data to upload to github    
#close.to_csv("../data/Close.csv")
#volume.to_csv("../data/Volume.csv")
```
